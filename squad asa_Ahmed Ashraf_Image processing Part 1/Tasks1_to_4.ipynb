```python
#Task 1

import cv2
import numpy as np
import matplotlib.pyplot as plt


def stackImages(scale, imgArray):
    rows = len(imgArray)
    cols = len(imgArray[0])
    rowsAvailable = isinstance(imgArray[0], list)
    width = imgArray[0][0].shape[1]
    height = imgArray[0][0].shape[0]
    if rowsAvailable:
        for x in range(0, rows):
            for y in range(0, cols):
                if imgArray[x][y].shape[:2] == imgArray[0][0].shape[:2]:
                    imgArray[x][y] = cv2.resize(imgArray[x][y], (0, 0), None, scale, scale)
                else:
                    imgArray[x][y] = cv2.resize(imgArray[x][y], (imgArray[0][0].shape[1], imgArray[0][0].shape[0]),
                                                None, scale, scale)
                if len(imgArray[x][y].shape) == 2: imgArray[x][y] = cv2.cvtColor(imgArray[x][y], cv2.COLOR_GRAY2BGR)
        imageBlank = np.zeros((height, width, 3), np.uint8)
        hor = [imageBlank] * rows
        hor_con = [imageBlank] * rows
        for x in range(0, rows):
            hor[x] = np.hstack(imgArray[x])
        ver = np.vstack(hor)
    else:
        for x in range(0, rows):
            if imgArray[x].shape[:2] == imgArray[0].shape[:2]:
                imgArray[x] = cv2.resize(imgArray[x], (0, 0), None, scale, scale)
            else:
                imgArray[x] = cv2.resize(imgArray[x], (imgArray[0].shape[1], imgArray[0].shape[0]), None, scale, scale)
            if len(imgArray[x].shape) == 2: imgArray[x] = cv2.cvtColor(imgArray[x], cv2.COLOR_GRAY2BGR)
        hor = np.hstack(imgArray)
        ver = hor
    return ver


image_height = 50
image_width = 50
number_of_color_channels = 3

img1 = np.full((image_height, image_width, number_of_color_channels), (255, 0, 0), dtype=np.uint8)
img2 = np.full((image_height, image_width, number_of_color_channels), (0, 255, 0), dtype=np.uint8)
img3 = np.full((image_height, image_width, number_of_color_channels), (0, 0, 255), dtype=np.uint8)
img4 = np.full((image_height, image_width, number_of_color_channels), (0, 0, 0), dtype=np.uint8)

stacked_image100 = stackImages(1, ([img1, img2], [img3, img4]))  # image resizing could have been done in this step  but is done in a later step according to the prerequistes of this task
final_image = cv2.resize(stacked_image100, (200, 200))

plt.imshow(final_image)
plt.show()
```

```python
#Task 2

#This program can't save videos if you changed from normal, grayscale, HSV to rotate or toggle all
#since they are of different dimentions
#so i added the save-video code again in the if conditions of the rotate and toggle all options along with a bool
#vriable that is set to true to start creating the file then set to false to prevent overwriting this file then start
#saving the video in the dimentions of the rotate/toggle_all video

import cv2
import numpy as np
import time as t

cap = cv2.VideoCapture(0)
rotate = False
HSV = False
grayscale = False
normal = False
toggle_all = False
save_vid = False
save_img = False
temp_r = True
temp_a = True

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

size = (frame_width, frame_height)
result = cv2.VideoWriter('filename.avi', cv2.VideoWriter_fourcc(*'MJPG'), 10, size)


def stackImages(scale, imgArray):
    rows = len(imgArray)
    cols = len(imgArray[0])
    rowsAvailable = isinstance(imgArray[0], list)
    width = imgArray[0][0].shape[1]
    height = imgArray[0][0].shape[0]
    if rowsAvailable:
        for x in range(0, rows):
            for y in range(0, cols):
                if imgArray[x][y].shape[:2] == imgArray[0][0].shape[:2]:
                    imgArray[x][y] = cv2.resize(imgArray[x][y], (0, 0), None, scale, scale)
                else:
                    imgArray[x][y] = cv2.resize(imgArray[x][y], (imgArray[0][0].shape[1], imgArray[0][0].shape[0]),
                                                None, scale, scale)
                if len(imgArray[x][y].shape) == 2: imgArray[x][y] = cv2.cvtColor(imgArray[x][y], cv2.COLOR_GRAY2BGR)
        imageBlank = np.zeros((height, width, 3), np.uint8)
        hor = [imageBlank] * rows
        hor_con = [imageBlank] * rows
        for x in range(0, rows):
            hor[x] = np.hstack(imgArray[x])
        ver = np.vstack(hor)
    else:
        for x in range(0, rows):
            if imgArray[x].shape[:2] == imgArray[0].shape[:2]:
                imgArray[x] = cv2.resize(imgArray[x], (0, 0), None, scale, scale)
            else:
                imgArray[x] = cv2.resize(imgArray[x], (imgArray[0].shape[1], imgArray[0].shape[0]), None, scale, scale)
            if len(imgArray[x].shape) == 2: imgArray[x] = cv2.cvtColor(imgArray[x], cv2.COLOR_GRAY2BGR)
        hor = np.hstack(imgArray)
        ver = hor
    return ver


while True:
    isTrue, img1 = cap.read()
    img = img1

    if cv2.waitKey(1) & 0xFF == ord('Q') or cv2.waitKey(1) & 0xFF == ord('q'):
        break

    if cv2.waitKey(1) & 0xFF == ord('R') or cv2.waitKey(1) & 0xFF == ord('r'):
        rotate = True

    if rotate:
        img = cv2.rotate(img1, cv2.ROTATE_90_CLOCKWISE)
        #the part dsecribed in the comments at the top
        if temp_r and cv2.waitKey(1) & 0xFF == ord('S') or cv2.waitKey(1) & 0xFF == ord('s'):
            frame_width = 480
            frame_height = 640
            size = (frame_width, frame_height)
            result = cv2.VideoWriter('filename.avi', cv2.VideoWriter_fourcc(*'MJPG'), 10, size)
            save_vid = True
            temp_r = False

        if save_vid:
            result.write(img)

    if cv2.waitKey(1) & 0xFF == ord('H') or cv2.waitKey(1) & 0xFF == ord('h'):
        HSV = True

    if HSV:
        img = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)

    if cv2.waitKey(1) & 0xFF == ord('G') or cv2.waitKey(1) & 0xFF == ord('g'):
        grayscale = True

    if grayscale:
        img = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

    if cv2.waitKey(1) & 0xFF == ord('Z') or cv2.waitKey(1) & 0xFF == ord('z'):
        normal = True

    if normal:
        img = img1

    if cv2.waitKey(1) & 0xFF == ord('X') or cv2.waitKey(1) & 0xFF == ord('x'):
        toggle_all = True

    if toggle_all:
        img2 = cv2.rotate(img1, cv2.ROTATE_90_CLOCKWISE)
        img3 = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
        img4 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        img = stackImages(1, ([img1, img2], [img3, img4]))
        # the part dsecribed in the comments at the top
        if temp_a and cv2.waitKey(1) & 0xFF == ord('S') or cv2.waitKey(1) & 0xFF == ord('s'):
            frame_width = 1280
            frame_height = 960
            size = (frame_width, frame_height)
            result = cv2.VideoWriter('filename.avi', cv2.VideoWriter_fourcc(*'MJPG'), 10, size)
            save_vid = True
            temp_a = False

        if save_vid:
            result.write(img)


    if cv2.waitKey(1) & 0xFF == ord('S') or cv2.waitKey(1) & 0xFF == ord('s'):
        save_vid = True
    if save_vid:
            result.write(img)

    if cv2.waitKey(1) & 0xFF == ord('C') or cv2.waitKey(1) & 0xFF == ord('c'):
        timestring = t.strftime("%Y%m%d-%H%M%S")
        cv2.imwrite("frame%s.jpg" % timestring, img)
    try:
        cv2.imshow("Webcam", img)
        pass
    except:
        print("webcam not available")
        break

cap.release()
result.release()
```

```python
#Task 3

# 48 x 20
# Each box is 10 px and represents 2 km^2
# Results are approxmated to the nearest box

import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

# creating background image
img = np.zeros((320, 560, 3), dtype='uint8')
img[:] = 255, 255, 255

# creating the gridlines
for i in range(40, 290, 10):
    cv2.line(img, (40, i), (520, i), (0, 0, 0), 1)

for i in range(40, 530, 10):
    cv2.line(img, (i, 40), (i, 280), (0, 0, 0), 1)

cv2.rectangle(img, (90, 130), (100, 140), (0, 0, 0), -1)

# Adding directions
cv2.putText(img, "N", (280, 30), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 0), 1)
cv2.putText(img, "S", (280, 310), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 0), 1)
cv2.putText(img, "W", (10, 160), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 0), 1)
cv2.putText(img, "E", (525, 160), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 0), 1)

angle = input("Enter the angle(in degree): ")
speed = input("Enter the speed(in m/s): ")
time_till_resurface = input("Enter the time till resurface(in hours): ")

time_in_seconds = int(time_till_resurface) * 60 * 60

distance = float(speed) * time_in_seconds  # in meters

if 0 < int(angle) < 90:



    angle_radians = int(angle) * math.pi / 180
    move_vertical = ((math.cos(angle_radians) * distance) / 1000) * 5
    # multiplied by 5 since each 10 pixels = 1 box and each box = 2 km
    move_horizontal = ((math.sin(angle_radians) * distance) / 1000) * 5
    location_horizontal = (95 + math.floor(move_horizontal)) + (10 - (95 + math.floor(move_horizontal)) % 10) + 5
    location_vertical = (135 - math.floor(move_vertical)) + (10 - (135 - math.floor(move_vertical)) % 10) + 5

    cv2.line(img, (95, 135), (location_horizontal, location_vertical), (0, 0, 0), 1)
    cv2.rectangle(img, (location_horizontal - 5, location_vertical - 5),
                  (location_horizontal + 5, location_vertical + 5), (255, 0, 0), -1)

elif 90 < int(angle) < 180:
    angle_radians = (180 - int(angle)) * math.pi / 180
    move_vertical = ((math.cos(angle_radians) * distance) / 1000) * 5
    move_horizontal = ((math.sin(angle_radians) * distance) / 1000) * 5
    location_horizontal = (95 + math.floor(move_horizontal)) + (10 - (95 + math.floor(move_horizontal)) % 10) +5
    location_vertical = (135 + math.floor(move_vertical)) + (10 - (135 + math.floor(move_vertical)) % 10)+5

    cv2.line(img, (95, 135), (location_horizontal, location_vertical), (0, 0, 0), 1)
    cv2.rectangle(img, (location_horizontal -5 , location_vertical -5),
                  (location_horizontal + 5, location_vertical + 5), (255, 0, 0), -1)

elif 180 < int(angle) < 270:
    angle_radians = (270 - int(angle)) * math.pi / 180
    move_vertical = ((math.sin(angle_radians) * distance) / 1000) * 5
    move_horizontal = ((math.cos(angle_radians) * distance) / 1000) * 5
    location_horizontal = (95 - math.floor(move_horizontal)) + (10 - (95 - math.floor(move_horizontal)) % 10) + 5
    location_vertical = (135 + math.floor(move_vertical)) + (10 - (135 + math.floor(move_vertical)) % 10) + 5

    cv2.line(img, (95, 135), (location_horizontal, location_vertical), (0, 0, 0), 1)
    cv2.rectangle(img, (location_horizontal - 5, location_vertical - 5),
                  (location_horizontal + 5, location_vertical + 5), (255, 0, 0), -1)

elif 270 < int(angle) < 360:
    angle_radians = (360 - int(angle)) * math.pi / 180
    move_vertical = ((math.cos(angle_radians) * distance) / 1000) * 5
    move_horizontal = ((math.sin(angle_radians) * distance) / 1000) * 5
    location_horizontal = (95 - math.floor(move_horizontal)) + (10 - (95 - math.floor(move_horizontal)) % 10) + 5
    location_vertical = (135 - math.floor(move_vertical)) + (10 - (135 - math.floor(move_vertical)) % 10) + 5

    cv2.line(img, (95, 135), (location_horizontal, location_vertical), (0, 0, 0), 1)
    cv2.rectangle(img, (location_horizontal - 5, location_vertical - 5),
                  (location_horizontal + 5, location_vertical + 5), (255, 0, 0), -1)


elif int(angle) == 0 or int(angle) == 360:

    move_vertical = (distance / 1000) * 5
    location_vertical = 135 - math.floor(move_vertical) + (10- (135 - math.floor(move_vertical))%10) + 5

    cv2.line(img, (95, 135), (95 , location_vertical), (0, 0, 0), 1)
    cv2.rectangle(img, (90, location_vertical - 5), (100, location_vertical + 5),
                  (255, 0, 0), -1)

elif int(angle) == 180:
    move_vertical = (distance / 1000) * 5
    location_vertical = 135 + math.floor(move_vertical) + (10 - (135 + math.floor(move_vertical)) % 10) + 5
    cv2.line(img, (95, 135), (95, location_vertical), (0, 0, 0), 1)
    cv2.rectangle(img, (95 - 5, location_vertical - 5), (95 + 5, location_vertical + 5),
                  (255, 0, 0), -1)

elif int(angle) == 90:
    move_horizontal = (distance / 1000) * 5
    location_horizontal = (95 + math.floor(move_horizontal)) + (10 - (95 + math.floor(move_horizontal)) % 10) + 5
    cv2.line(img, (95, 135), (location_horizontal, 135), (0, 0, 0), 1)
    cv2.rectangle(img, (location_horizontal - 5, 135 - 5), (location_horizontal + 5, 135 + 5),
                  (255, 0, 0), -1)

elif int(angle) == 270:
    move_horizontal = (distance / 1000) * 5
    location_horizontal = (95 - math.floor(move_horizontal)) + (10 - (95 - math.floor(move_horizontal)) % 10) + 5
    cv2.line(img, (95, 135), (location_horizontal, 135), (0, 0, 0), 1)
    cv2.rectangle(img, (location_horizontal - 5, 135 - 5), (location_horizontal + 5, 135 + 5),
                  (255, 0, 0), -1)

img_plot = plt.imshow(img)
plt.show()

```
